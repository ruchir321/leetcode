# 1757. Recyclable and Low Fat Products

```python
import pandas as pd

data = {
    'product_id': [0, 1, 2, 3, 4],
    'low_fats': ['Y', 'Y', 'N', 'Y', 'N'],
    'recyclable': ['N', 'Y', 'Y', 'Y', 'N']
}

products = pd.DataFrame(data)

# Set data types according to schema
products['product_id'] = products['product_id'].astype(int)
products['low_fats'] = pd.Categorical(products['low_fats'], categories=['Y', 'N'])
products['recyclable'] = pd.Categorical(products['recyclable'], categories=['Y', 'N'])

print(products)
print("\n--- Data Types ---")
print(products.dtypes)
```

```python
product_filter = (products["low_fats"] == "Y") & (products["recyclable"] == "Y")
```

```python
type(pd.DataFrame(products[product_filter]["product_id"]))
```
